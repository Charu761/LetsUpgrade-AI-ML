### Inference from the analysis:




### Correlation

```python
from scipy.stats import pearsonr

#convert Attrition values yes and no as 0 and 1 to be able to correlate and calculate the r value
#dataset.Attrition = dataset.Attrition.replace('No', 0)
#dataset.Attrition = dataset.Attrition.replace('Yes', 1)

from sklearn import preprocessing
label_encoder=preprocessing.LabelBinarizer()
dataset['Attrition']=label_encoder.fit_transform(dataset["Attrition"])

dataset.head()
```

### 1. Correlation between Attrition and Age

```python
stats,p=pearsonr(dataset.Attrition, dataset.Age)
print(stats,p)
```

#### Here in the above test we found the correlation between the age and attrition and we got the p-value is less than 0.005 so (h0) is rejected and (ha) is accepted and we can conclude that there is significant correlation between the age and attrition

### 2. Correlation between Attrition and MonthlyIncome

```python
stats,p=pearsonr(dataset.Attrition, dataset.MonthlyIncome)
print(stats,p)
```

#### Here in the above test we found the correlation between the MonthlyIncome and attrition and we got the p-value is less than 0.005 so (h0) is rejected and (ha) is accepted and we can conclude that there is significant correlation between the MonthlyIncome and attrition

###  3. Next to find the correlation between the Attrition and DistanceFromHome

```python
stats,p=pearsonr(dataset.Attrition, dataset.DistanceFromHome)
print(stats,p)
```

#### The p-value is greater than 0.005 so (h0) is accepted and we can conclude that there is no correlation between the DistanceFromHome and Attrition

### Statistical Tests (Mann-Whitney)

```python
dataset_yes = dataset[dataset['Attrition']==1]
dataset_no  = dataset[dataset['Attrition']==0]
```

```python
dataset_yes.head()
```

```python
dataset_no.head()
```

```python
from scipy.stats import mannwhitneyu

stats,p=mannwhitneyu(dataset_yes.Education,dataset_no.Education)
print(stats,p)
```

### Above test was conducted to test:

#### h0 : there is no significant difference in Education between the attrition group and non attrition group

#### ha : there is a significant difference in Education between the attrition group and non attrition group

####  Since the p-value is greater than 0.05 we can accept the h0 hypothesis.

```python
stats,p=mannwhitneyu(dataset_yes.Age,dataset_no.Age)
print(stats,p)
```

### Above test was conducted to test:

#### h0 : there is no significant difference in Age between the attrition group and non attrition group

#### ha : there is a significant difference in Age between the attrition group and non attrition group

####  Since the p-value is less than 0.05 we can say that there is Age factor in the Attrition.

```python
stats,p=mannwhitneyu(dataset_yes.DistanceFromHome,dataset_no.DistanceFromHome)
print(stats,p)
```
